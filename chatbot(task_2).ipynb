```python
print("Chatbot: Hello! Type 'bye' to exit.")

while True:
    user = input("You: ").lower()

    if user == "bye":
        print("Chatbot: Goodbye!")
        break

    elif "hello" in user or "hi" in user:
        print("Chatbot: Hi there! How can I help you?")

    elif "how are you" in user:
        print("Chatbot: I'm just code, but I'm doing great!")

    elif "your name" in user:
        print("Chatbot: I'm a simple rule-based chatbot.")

    elif "help" in user:
        print("Chatbot: I can answer simple questions. Try asking about me!")

    elif "weather" in user:
        print("Chatbot: I can't check real weather yet, but I hope it's sunny!")

    else:
        print("Chatbot: Sorry, I don't understand that.")
```
